```
import torch
from torch import autocast
from diffusers import StableDiffusionPipeline
```

```
pipeline = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4")
pipeline = pipeline.to("cuda")
```

```
length = float(input("Enter kitchen length (in meters): "))
breadth = float(input("Enter kitchen breadth (in meters): "))
height = float(input("Enter kitchen height (in meters): "))
num_chairs = int(input("Enter the number of chairs: "))
num_tables = int(input("Enter the number of tables: "))
```

```
prompt = f"A beautiful kitchen design with dimensions {length}x{breadth}x{height} meters, containing {num_chairs} chairs and {num_tables} tables."
```

```
generator = torch.manual_seed(42)
with autocast("cuda"):
    image = pipeline(prompt).images[0]
```

```
image.save("kitchen_design4.png")
image.show()
```
